```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the grid
grid = np.zeros((50, 50))

# Defining the locations
nest_location = (25, 5)
strong_feeder_location = (10, 40)
weak_feeder_location = (40, 40)

#1 for the nest, 2 for the strong feeder, and 3 for the weak feeder
grid[nest_location] = 1
grid[strong_feeder_location] = 2
grid[weak_feeder_location] = 3

fig, ax = plt.subplots(figsize=(8,8))
cax = ax.matshow(grid, cmap=plt.cm.get_cmap('Greys', 4), origin='lower')
ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=0.1)
ax.set_xticks(np.arange(0, 50, 1))
ax.set_yticks(np.arange(0, 50, 1))

ax.tick_params(axis='both', which='both', length=0)
ax.set_xticklabels([])
ax.set_yticklabels([])

ax.text(nest_location[1], 50 - nest_location[0] - 1, f'Nest ({nest_location[0]}, {nest_location[1]})', ha='center', va='center', color='r')
ax.text(strong_feeder_location[1], 50 - strong_feeder_location[0] - 1, f'Strong Feeder ({strong_feeder_location[0]}, {strong_feeder_location[1]})', ha='center', va='center', color='r')
ax.text(weak_feeder_location[1], 50 - weak_feeder_location[0] - 1, f'Weak Feeder ({weak_feeder_location[0]}, {weak_feeder_location[1]})', ha='center', va='center', color='r')

plt.show()
```
